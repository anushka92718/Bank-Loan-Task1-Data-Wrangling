```python
import pandas as pd
df = pd.read_csv('financial_loan.csv')
print("Initial Data Overview:")
df.info()
df.head()
```

```python
# 1. Check if 'id' is unique and has no missing values
if df['id'].is_unique and not df['id'].isnull().any():
    print("Success: 'id' is unique and contains no nulls. Setting as Index (Primary Key).")
    df.set_index('id', inplace=True)
else:
    # If there are duplicates, this will help you find them
    duplicates = df[df.duplicated('id')]
    print(f"Error: Found {len(duplicates)} duplicate IDs. Please check data.")

df.head()
```

```python
# 2. Identify all text (object) columns
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    # Convert to string and remove leading/trailing whitespace
    df[col] = df[col].astype(str).str.strip()
    
    # Truncate to 100 characters (this handles your 'varchar(100)' requirement)
    df[col] = df[col].str.slice(0, 100)
    
    # Convert empty strings or 'nan' strings back to actual NaN for SQL compatibility
    df[col] = df[col].replace(['nan', 'None', ''], pd.NA)

print(f"Cleaned {len(text_cols)} text columns to match varchar(100) requirements.")
```

```python
# 3. Ensure total_payment is an integer
# Note: if there are missing values (NaN), it must be 'Int64' (nullable) or fill the NaNs first
df['total_payment'] = pd.to_numeric(df['total_payment'], errors='coerce').fillna(0).astype(int)

print("Check Datatype:")
print(df['total_payment'].dtype)
```

```python
# Save the final cleaned dataset
output_filename = 'cleaned_financial_loan.csv'
df.to_csv(output_filename)

print(f"Cleaning complete! Your file '{output_filename}' is ready for download.")
```
